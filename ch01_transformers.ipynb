```python
# Load model directly

from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained(
    "openai-community/gpt2",
    token=hf_token
)

model = AutoModelForCausalLM.from_pretrained(
    "openai-community/gpt2",
    token=hf_token
)
```

## tokenizer

```python
# Word to tokenize
text = "unbelivable"

# Convert the text into token IDs and return them as PyTorch tensors
tokenized_output = tokenizer(text, return_tensors="pt")

# Display the complete tokenizer output dictionary
print("Tokenizer Output:")
print(tokenized_output)

# Extract and display the token IDs
token_ids = tokenized_output["input_ids"][0]
print("\nToken IDs:")
print(token_ids)

# Convert token IDs back to text and display the decoded result
decoded_text = tokenizer.decode(token_ids)
print("\nDecoded Text:")
print(decoded_text)
```

```python
for token_id in token_ids:
  print(f"{token_id} -> {tokenizer.decode([token_id])}")
```

```python
input_text = "I like machine learning to"

for i in range(6):
  inputs = tokenizer(input_text, return_tensors="pt").input_ids
  out = model(inputs).logits[0,-1]
  out_word = tokenizer.decode(out.argmax())
  print(out_word)
  input_text = input_text + " " + out_word

input_text
```

```python
import torch

input_text = "I like machine learning to"

for i in range(6):
    inputs = tokenizer(input_text, return_tensors="pt").input_ids

    logits = model(inputs).logits
    next_token_logits = logits[0, -1, :]   # last position

    next_token_id = torch.argmax(next_token_logits).item()
    next_token = tokenizer.decode([next_token_id])

    input_text += next_token

print(input_text)
```

```python
input_text = "I like machine learning to"

k = 10  # consider top 10 candidates
for i in range(6):
    inputs = tokenizer(input_text, return_tensors="pt").input_ids
    logits = model(inputs).logits
    next_token_logits = logits[0, -1, :]
    # get top-k tokens
    top_values, top_indices = torch.topk(next_token_logits, k)
    # convert logits → probabilities
    probs = torch.softmax(top_values, dim=-1)
    # sample from top-k instead of picking max
    sampled_index = torch.multinomial(probs, num_samples=1)
    next_token_id = top_indices[sampled_index].item()
    next_token = tokenizer.decode([next_token_id])
    input_text += next_token
print(input_text)
```
