```python
from google.colab import userdata
hf_token = userdata.get('HUGGING_FACE_API')
```

```python
# Use Pipeline as a high-level helper

from transformers import pipeline

pipe = pipeline("text-generation",model="openai-community/gpt2",token=hf_token)
```

```python
prompt = "Explain machine learning in one sentence"
result = pipe(prompt)
print(result)
```

```python
print(result[0]['generated_text'])
```

```python
# Load model directly

from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained(
    "openai-community/gpt2",
    token=hf_token
)

model = AutoModelForCausalLM.from_pretrained(
    "openai-community/gpt2",
    token=hf_token
)
```

## tokenizer

```python
# Word to tokenize
text = "unbelivable"

# Convert the text into token IDs and return them as PyTorch tensors
tokenized_output = tokenizer(text, return_tensors="pt")

# Display the complete tokenizer output dictionary
print("Tokenizer Output:")
print(tokenized_output)

# Extract and display the token IDs
token_ids = tokenized_output["input_ids"][0]
print("\nToken IDs:")
print(token_ids)

# Convert token IDs back to text and display the decoded result
decoded_text = tokenizer.decode(token_ids)
print("\nDecoded Text:")
print(decoded_text)
```

```python
for token_id in token_ids:
  print(f"{token_id} -> {tokenizer.decode([token_id])}")
```

## AutoModelForCausalLM

```python
input_text = "The capital of France is"
inputs = tokenizer(input_text, return_tensors="pt").input_ids

print(inputs)

model(inputs)
```
